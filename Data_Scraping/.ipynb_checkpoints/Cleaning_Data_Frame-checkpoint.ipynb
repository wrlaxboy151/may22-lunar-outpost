```python
# import packages that we need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
sns.set_style("whitegrid")
```

```python
# read in pikl file 
movie_data = pd.read_pickle('Data_from_imdb.pkl')
```

```python
movie_data.head()
```

```python
# adjust column titles 
original_columns = movie_data.columns
```

```python
original_columns
```

```python
new_columns = {'imdb id': 'imdb_id','origingal air date':'original_air_date', 'color info':'color_info', 
               'box office':'box_office', 'production company':'production_company'}
```

```python
updated_movie_data = movie_data.rename(new_columns,axis='columns')
```

```python
updated_movie_data.columns
```

```python
# read in nominee file 
nominees = pd.read_csv('Data/df_of_nominees.csv',dtype={'imdb_id':'str', 'is_nominee':'int'})
```

```python
nominees.head()
```

```python
is_nominee = nominees['is_nominee'].values
```

```python
# append is_nominee to dataframe
updated_movie_data['is_nominee'] = is_nominee
```

```python
updated_movie_data.head()
```

```python
# print to pkl format 
updated_movie_data.to_pickle('Data/df_of_imdb_data.pkl')
new_movie_data = pd.read_pickle('Data/df_of_imdb_data.pkl')
```
