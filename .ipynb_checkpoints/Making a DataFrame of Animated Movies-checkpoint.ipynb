```python
titles=[]
casts=[]
years=[]
production_companies=[]
countries=[]
distributors=[]
certificates=[]
genres=[]
original_air_date=[]
rating=[]
votes=[]

#movies2=pd.DataFrame({'imdb_id': imdb_ids[0:3]})

for id in movies['imdb_id']:
    movie=ia.get_movie(id)
    if 'title' in movie.keys():
        titles.append(movie['title'])
    else:
        titles.append('')
    if 'cast' in movie.keys():
        casts.append(movie['cast'])
    else:
        casts.append('')
    if 'year' in movie.keys():
        years.append(movie['year'])
    else:
        years.append('')
    if 'production companies' in movie.keys():
        production_companies.append(movie['production companies'])
    else:
        production_companies.append('')
    if 'countries' in movie.keys():
        countries.append(movie['countries'])
    else:
        countries.append('')
    if 'distributors' in movie.keys():
        distributors.append(movie['distributors'])
    else:
        distributors.append('')
    if 'certificates' in movie.keys():
        certificates.append(movie['certificates'])
    else:
        certificates.append('')
    if 'genres' in movie.keys():
        genres.append(movie['genres'])
    else:
        genres.append('')
    if 'original air date' in movie.keys():
        original_air_date.append(movie['original air date'])
    else:
        original_air_date.append('')
    if 'rating' in movie.keys():
        rating.append(movie['rating'])
    else:
        rating.append('')
    if 'votes' in movie.keys():
        votes.append(movie['votes'])
    else:
        votes.append('')
    time.sleep(1)

movies['title']=titles
movies['cast']=casts
movies['year']=years
movies['production_companies']=production_companies
movies['countries']=countries
movies['distributors']=distributors
movies['certificates']=certificates
movies['genres']=genres
movies['air_date']=original_air_date
movies['imdb_rating']=rating
movies['imdb_votes']=votes

movies.to_csv("df_of_movies.csv")
```
