Importing Stuff

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime, timedelta
from seaborn import set_style
from sklearn.metrics import mean_squared_error

set_style("whitegrid")
```

Downloading Data Files and Sorting by Year

```python
movie_data=pd.read_csv("Data_from_imdb.csv")
print(movie_data.columns)

#Adding Missing Year
movie_data.at[3107,'year']="2019"

#Sort the dataframe by year
sorted_movie_data = movie_data.sort_values(by = ['year']).reset_index()
```

Adding a column to the data frame that indicates being nominated or not. 

```python
#Import Nomination Ids
nominee_ids = pd.read_csv("nominee_ids.csv")

print(nominee_ids.head())

sorted_movie_data["nom_status"] = sorted_movie_data["imdb id"].isin(nominee_ids.imdb_id).astype(int)
print(sorted_movie_data["nom_status"].sample(100))
```

Plotting Run Time over Release Year

```python
plt.figure(figsize=(20,10))

plt.plot(sorted_movie_data.year, sorted_movie_data.runtimes)

plt.xlabel("Release Year", fontsize=18)
plt.ylabel("Run Time", fontsize=18)

plt.xticks(fontsize=12)
plt.yticks(fontsize=1)

plt.show()
```

Plot of Nomination Status Over Time (i.e. number of nominees). It makes sense that this is uniform, because (I think) there are the same number of nominations each year. 

```python
plt.figure(figsize=(20,10))

plt.plot(sorted_movie_data.year, sorted_movie_data.nom_status)

plt.xlabel("Release Year", fontsize=18)
plt.ylabel("Nomination Status", fontsize=18)

plt.xticks(fontsize=12)
plt.yticks(fontsize=1)

plt.show()
```
