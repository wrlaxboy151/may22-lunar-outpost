```python
temp = pd.read_pickle('df_of_imdb_data.pkl')
```

```python
temp.head()
```

```python
nom_directors = temp.loc[(temp['is_nominee']==1) & (temp['year']==2001)]['directors'].explode()
```

```python
test = []
for year in range(2001, 2021, 1):
    nom_directors = temp.loc[(temp['is_nominee']==1) & (temp['year']==year)]['directors'].explode()
    for director in nom_directors:
        for future_year in range(year+1, 2022, 1):
            future_films = temp.loc[(temp['year']==future_year)]
            for future_director in future_films['directors']:
                if director in future_director:
                    test.append(1)
                else:
                    test.append(0)
```

```python
for year in range(2021, 2001, -1):
    films_by_year = temp.loc[temp['year']==year]
    for directors in films_by_year['directors']:
        print(directors)
```
